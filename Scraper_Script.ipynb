```python
result_list = []
```

```python
all_li = driver.find_elements(By.XPATH, '//div[@class="company"]')
all_li
```

```python
for i in all_li:
    try:
        ActionChains(driver).move_to_element(i).click(i).perform()
        company_name = driver.find_element(By.XPATH, '//h1[@class="coSmallHeader"]').text
        company_address = driver.find_element(By.XPATH, '//*[@id="address_div"]/table/tbody/tr[1]/td').text
        company_website = driver.find_element(By.XPATH, '//a[@id="ctl00_MainContent_hlWebsite"]').get_attribute('href')
        result = {
            'company_name': company_name,
            'company_address': company_address,
            'company_website': company_website
        }
        result_list.append(result)
        driver.implicitly_wait(3)
    except (NoSuchElementException, StaleElementReferenceException, ElementNotInteractableException):
        pass
```
